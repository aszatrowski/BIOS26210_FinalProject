```python
#### NON-MEMORY MODEL V1

import numpy as np #package for work with arrays and matrices
import matplotlib.pyplot as plt #package with plotting capabilities

dt = 1 #set the time step
Tmax = 400 #set the maximum time duration
ts = np.arange(0,Tmax, dt)
Niter = int(np.ceil(Tmax/dt)) #determine the number of iterations
x = np.zeros(Niter) #preallocate the solution array
x[0] = 0 #set the initial value


A = 270
B1 = 2441
B2 = 90.02
T1 = 19.6
T2 = 200.5
DImin = 53.5
n = 2          #make it 1 or 2 maybe play around with other values later on

for i in range(Niter-1):
    
    x[i+1] = A - B1*np.exp((x[i]-n*i)/T1) - B2*np.exp((x[i]-n*i)/T2)
    if n*i - x[i+1] >= DImin:
        x[i+1] = 0                            #made the zero to get them out of the way visually, 
plt.scatter(ts,x, label='numeric solution')



# y = A - B1*np.exp((x[0]-n*ts)/T1) - B2*np.exp((x(0)-n*ts)/T2)
# plt.plot(ts,y, label='analytic') #plot the solution on the same figure
plt.xlabel('time')  
plt.ylabel('solution') 
plt.title('APD v. ts') 
plt.legend()
plt.show()
```
